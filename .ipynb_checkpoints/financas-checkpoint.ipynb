```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
```

```python
# carregar dados da empresa TAESA
taesa = yf.Ticker('TAEE11.SA')

# criar pandas DataFrame
df = taesa.history(period='1mo')

# classificar desde a primeira data
df.sort_values(by='Date')
```

```python
# quantidade de linhas
len(df)
```

```python
# setar estilo
sns.set_theme(style='darkgrid')

#tamanho da figura
plt.figure(figsize=(25,10))

# plotar o grfico
ax = sns.lineplot(x=df.index.values, y=df['Close'])

# setar os labels dos eixos
ax.set_ylabel('Valores - BRL', fontsize=20, labelpad=40)
ax.set_xlabel('Perodo', fontsize=20, labelpad=40)

# setar os labels do ttulo
ax.set_title('TAESA - Transmissora Alianca Energia Eletrica SA Unit', fontsize = 22, verticalalignment='baseline', pad=40)

# set parametros dos eixos
ax.tick_params(axis='y', labelsize=15, pad=8)
ax.tick_params(axis='x', labelsize=15, pad=8)

#formatar data para o eixo X
date_form = DateFormatter('%d/%m/%Y')
ax.xaxis.set_major_formatter(date_form)


plt.show()
```
